```python
df_paint=scaled_df[['TYPE_ATTENTION', 'DETAIL_ATTENTION', 'SEQ_ATTENTION', 'SEQ_PATIENT',
       'SEQ_ATTEN_PATIENT', 'ROOM_ATTENTION', 'MAIN_SPECIALITY', 'SEQ_DOCTOR',target]].copy()

sns.histplot(data=scaled_df, x=target)
```

```python
# Matriz de Correlacin de variables
corr = df_paint.corr(method="pearson")
sns.heatmap(corr, cmap='coolwarm', annot=True, fmt='.6f')
#plt.show()
```

```python
corr_matrix = scaled_df.corr(method='pearson')
corr_matrix
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
# Heatmap matriz de correlaciones
# ==============================================================================
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(5, 5))

sns.heatmap(
    corr_matrix,
    annot     = True,
    cbar      = False,
    annot_kws = {"size": 5},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 220, n=200),
    square    = True,
    ax        = ax
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)

ax.tick_params(labelsize = 10)
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
# Graficamos la variable Class separada por los distintos atributos
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12, 8))
for i, variable in enumerate(x_columns[:-1]):
    row = i % 2
    col = i // 2
    sns.countplot(data=df, x=target,hue=variable, ax=axes[row][col])
    axes[row][col].set_title(f"Variable objetivo vs el atributo {variable}")
plt.tight_layout()
plt.show()
```
